```python
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()

display(iris['data'][:,3]) # petal width
display(iris['target'])
```

```python
X = iris['data'][:,2] # petal width
y = (iris['target'] == 2).astype(int) # True=1 for virginica, False=0 for all others

display(X)
display(y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.xlim([2.5,7.5])
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X.reshape(-1,1),y)

X_new = np.linspace(0,10,1000).reshape(-1,1)
y_prob = log_reg.predict_proba(X_new)

plt.scatter(X,y)
plt.plot(X_new, y_prob[:,1], 'g-', label='Virginica')
plt.plot(X_new, y_prob[:,0], 'b--', label='not Virginica')
plt.xlim([2.5,7.5])
plt.legend()
```

```python
X = iris['data'][:,2:] # petal length and petal width
y = (iris['target'] == 2).astype(int) # True=1 for virginica, False=0 for all others
```

```python
X
```

```python
log_reg = LogisticRegression(multi_class='multinomial',solver ='newton-cg')
log_reg.fit(X,y)

X_test,Y_test = np.meshgrid(np.arange(0, 10, 0.1), np.arange(0, 3, 0.01))
X_test = np.array([X_test.flatten(),Y_test.flatten()]).T
```

```python
X_test
```

```python
y_probs = log_reg.predict_proba(X_test)

y_probs
```
